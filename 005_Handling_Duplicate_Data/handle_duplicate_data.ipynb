# Detect and Drop the Duplicate dummy Data in Machine Learning

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Name": ["a", "b", "c", "d", "a", "c"],
                       "Eng": [8,7,5,8,8,5],
                       "Hindi": [2,3,4,5,2,6]})
df
```

```python
#  Only Tell true if the entire row content is same for more than one row
# df.duplicated()
# df["duplicate"] = df.duplicated()
df["duplicate"] = df.duplicated()
df
```

```python
df.drop_duplicates(keep = "first", inplace = True)
df
```
