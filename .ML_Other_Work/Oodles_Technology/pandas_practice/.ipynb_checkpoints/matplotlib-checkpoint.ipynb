```python
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.xlabel("Marks")
plt.ylabel("Category")
plt.plot(x,y)
plt.title("FIrst")

plt.subplot(1, 2, 2)
plt.xlabel("Marks")
plt.ylabel("Category")
plt.plot(x1,y1)
plt.title("Second")
# plt.legend()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.xlabel("Marks")
plt.ylabel("Category")
plt.bar(x,y)
plt.title("FIrst")

plt.subplot(1, 2, 2)
plt.xlabel("Marks")
plt.ylabel("Category")
plt.bar(x1,y1)
plt.title("Second")
# plt.legend()
```

```python
dataset = pd.DataFrame({'Payment': [90,34,76,56,86,97], 
                        'Country': ["India", "USA", "Canada", "Demark", "USA", "UK"],
                        'PaymentMode': ["UPI", "Cash", "Credit", "Cash", "UPI", "Credit"]})
dgrouped = dataset.groupby(['PaymentMode', 'Country'])['Payment'].agg({'mean', 'sum'})
dgrouped
```
