# OrdinalEncoding Using the Skicit Library

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Size": ["s", "m", "l", "xl", "s", "xl", "l", "m"]})
df
```

```python
ord_data = [["s", "m", "l" ,"xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = ord_data)
oe.fit(df[["Size"]])
```

```python
oe.transform(df[["Size"]])
df
```

# Ordinal Encoding using the Map Function

```python
ord_data1 = {"s": 0, "m": 1, "l" : 2, "xl": 3}
ord_data1
```

```python
df["Size_en_map"] = df["Size"].map(ord_data1)
df
```

# Working With the Large Dataset means the CSV File

```python
dataset = pd.read_csv("loan.csv")
dataset.head(10)
```

```python
dataset["Property_Area"].unique()
```

```python
dataset["Property_Area"].fillna(dataset["Property_Area"].mode()[0], inplace = True)
dataset
```

```python
en_data_ord = [["Rural" , "Semiurban", "Urban"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = en_data_ord)
dataset["Property_Area"] = oe.fit_transform(dataset[["Property_Area"]])
```
