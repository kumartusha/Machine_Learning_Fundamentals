# Wine Quality Prediction

```python
#  Import the required Libraries and Requirements
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Data Collection Steps. 
dataset = pd.read_csv("winequality.csv")
dataset.head()
```

```python
#  Check the rows and columns for the dataset.
print(dataset.shape)
```

```python
#  Entire DataSet.
dataset    # we have all the numeric values into the dataset.
```

```python
#  Data Preprocessing .( Missing Value )
dataset.isnull().sum()  # Here we dont have the null value in the entire dataset. (It is great for us )
```

## Data Analysis and the Visualization Part

```python
#  Statistical functions for the dataset.
dataset.describe()
```

```python
#  Number of values for each Quality
sns.catplot(x = "quality" , data = dataset, kind='count', palette='pastel')

# dataset["quality"].value_counts()
```

```python
# Compare the Volatile Acidity and Quality.
plot = plt.figure(figsize=(5,5))
sns.barplot(x = 'quality', y = 'volatile acidity', data=dataset, palette='pastel')
plt.show()

#  Citric Acid is inversely proportional to the quality
```

```python
# Use the Heatmap for showing the relation between the ciolumn to each other
plt.figure(figsize=(8,8))
sns.heatmap(correlation, annot=True, fmt='.1f', annot_kws={'size': 8}, cmap='Blues')
plt.show()
```

## Data Preprocessing Part

```python
#  Seperate the training and the testing data into the dependent and the independent variable.
x = dataset.drop("quality", axis = 1)
y = dataset["quality"]
```

## Label Classification or Label Encoding

```python
# We separete the quality of the data into the label formization.
y = dataset["quality"].apply(lambda y_value: 1 if y_value >= 7 else 0)
```

## Train and Test Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

## Model Training :- Random Forest Classifier

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

## Evaluation of the Model Performance

```python
#  Accuracy for the training data.
x_train_pred = model.predict(x_train)
x_train_accu = accuracy_score(x_train_pred, y_train)
print("The training accuracy are:- " , x_train_accu * 100)

#  Accuracy for the testing data.
x_test_pred = model.predict(x_test)
x_test_accu = accuracy_score(x_test_pred, y_test)
print("The testing accuracy are:- " , x_test_accu * 100)
```

## Building the Predictive System for the Model.

```python
#  Here we have the list of the new input data.
# input_data = (11.2,0.28,0.56,1.9,0.075,17.0,60.0,0.998,3.16,0.58,9.8)
# input_data = (7.3,0.65,0.0,1.2,0.065,15.0,21.0,0.9946,3.39,0.47,10.0)
input_data = (9.6,0.32,0.47,1.4,0.055999999999999994,9.0,24.0,0.99695,3.22,0.82,10.3)

#  we need to convert this list or tuple into the numpy array.
input_data_numpy = np.asarray(input_data)

#  for giving the multiple features and to give to the model we need to convert into the 2D array.
shaped_array = input_data_numpy.reshape(1,-1)

prediction = model.predict(shaped_array)

if(prediction[0] == 0):
    print("The wine quality is Bad")
else:
    print("The wine Quality is Good")
```
