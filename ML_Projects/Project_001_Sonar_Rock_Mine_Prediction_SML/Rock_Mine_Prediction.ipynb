# SONAR ROCK VS MINE PREDICTION

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#  Import the data using the pandas..
dataset = pd.read_csv("sonar.csv", header= None)
dataset
```

```python
dataset.describe()   # check the statistical methods for all the rows and columns.
```

```python
dataset.isnull().sum()  # check the null Value..
```

```python
dataset[60].value_counts()   # count the values from the dataset
```

```python
dataset.groupby(60).mean()   # Grouping of mean.
```

```python
# divide the dataset into the dependent and the independent variable..
x = dataset.iloc[:, : -1]
y = dataset.iloc[:, -1]
```

Training and Testing Data


```python
# Divide the data into the training and the testing
# Stratify is used to split the data based on the Label/Target Column
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, stratify = y, random_state = 2)
```

```python
# Here we have check the size of the training and the testing data.
print(x.shape, x_train.shape, x_test.shape)
```

```python
print(x_train)
print(y_train)
```

Model Training

```python
model = LogisticRegression()
```

```python
# training the logistic regression model with the training data.
model.fit(x_train, y_train)
```

Model Accuracy.

```python
#  Here we find the accuracy of the training data.
x_pred = model.predict(x_train)
train_data_accuracy = accuracy_score(x_pred, y_train)


#  Here we find the accuracy of the testing data.
x_pred2 = model.predict(x_test)
test_data_accuracy = accuracy_score(x_pred2, y_test)
```

```python
print(train_data_accuracy * 100)
print(test_data_accuracy * 100)
```
