```python
#  Drop the Highly correlated Feature.
#  Multicollinearity
dataset.drop("id", axis = 1, inplace=True)
```

##  Feature Selection 

```python
# 32 Features reduces to the 23 it is called the Feature Selecting.

corr_matrix = dataset.corr().abs()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
tri_dataset = corr_matrix.mask(mask)

print(len(tri_dataset.columns))
# If any value is greater than the 0.92% we simply removed that high correlated data.
t_drop = [x for x in tri_dataset.columns if any(tri_dataset[x] > 0.92)]
print(t_drop)

#  we are drop those features from our dataset whose value are greater than the 0.92
dataset = dataset.drop(t_drop, axis = 1)
print(dataset.shape)
```

## Split into the dependent and the independent Variable.

```python
X = dataset.drop("diagnosis", axis = 1)
Y = dataset["diagnosis"]
```

```python
## Split the data into the training and the testing data.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```
