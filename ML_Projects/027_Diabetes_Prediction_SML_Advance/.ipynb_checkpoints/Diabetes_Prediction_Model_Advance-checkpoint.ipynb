```python
#  But in this dataset here we have the catch like we have the 0 value rather than the empty cell in the dataset.
#  so we need to replace all the zeroes from the NaN
dataset[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']] = dataset[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']].replace(0, np.NaN)
```

```python
import missingno as miso
```

```python
#  we need to fill the missing cells in the dataset with the median Value.
dataset.groupby("Outcome").agg({"Pregnancies": ["mean","median"]})
```

```python
#  Fill the median Value with the help of the Outcome Column.
def median_target(var):    # Takes the features from the dataset
    temp = dataset[dataset[var].notnull()]
    temp = temp[[var, "Outcome"]].groupby(["Outcome"])[[var]].median().reset_index()
    return temp
```

```python
columns = dataset.columns
columns = columns.drop("Outcome")
for i in columns:
    median_target(i)
    dataset.loc[(dataset["Outcome"] == 0) & (dataset[i].isnull()), i] = median_target(i)[i][0]
    dataset.loc[(dataset["Outcome"] == 1) & (dataset[i].isnull()), i] = median_target(i)[i][1]
```

```python
# Outlier Detection
# IQR + Q1
# 50%
outlier_list = []
for i in dataset:
    Q1 = dataset[i].quantile(0.25)
    Q3 = dataset[i].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    if dataset[(dataset[i] > upper)].any(axis = None):
        print(i, "Yes")
        outlier_list.append(i)
    else:
        print(i, "No")
```

```python
#  we need to remove all the outliers from the columns.
# for i in outlier_list:
Q1 = dataset.Insulin.quantile(0.25)
Q2 = dataset.Insulin.quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

dataset.loc[dataset["Insulin"] > upper, "Insulin"] = upper
```

## Feature Engineering (Modify or add new Feature is called the feature Engineering)
We need to understand our data efficiently like which is need to be updated and categorize in the new feature
like we have the BMI index with different parameter.
In case of the Insulin we also modify it.

```python
NewBMI = pd.Series(["Underweight", "Normal", "Overweight", "Obesity 1", "Obesity 2", "Obesity 3"], dtype="category")
dataset["NewBMI"] = NewBMI
dataset.loc[dataset["BMI"] < 18.5, "NewBMI"] = NewBMI[0]
dataset.loc[(dataset["BMI"] > 18.5) & dataset["BMI"] <= 24.9, "NewBMI"] = NewBMI[1]
dataset.loc[(dataset["BMI"] > 24.9) & dataset["BMI"] <= 29.9, "NewBMI"] = NewBMI[2]
dataset.loc[(dataset["BMI"] > 29.9) & dataset["BMI"] <= 34.9, "NewBMI"] = NewBMI[3]
dataset.loc[(dataset["BMI"] > 34.9) & dataset["BMI"] <= 39.9, "NewBMI"] = NewBMI[4]
dataset.loc[dataset["BMI"] > 39.9, "NewBMI"] = NewBMI[5]
```

```python
dataset.head()
```

```python
# If insulin >= 16 & inculin <= 166  -> Normal
def set_insulin(row):
    if row["Insulin"] >= 16 and row["Insulin"] <= 166:
        return "Normal"
    else:
        return "Abnormal"
```

```python
dataset = dataset.assign(NewInsulinScore = dataset.apply(set_insulin, axis = 1))
```
