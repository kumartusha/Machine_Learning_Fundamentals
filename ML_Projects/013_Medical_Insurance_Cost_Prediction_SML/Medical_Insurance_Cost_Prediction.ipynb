## Label Encoding into the dataset.

```python
dataset = dataset.replace({"sex": {"male" : 1, "female": 0}, "smoker": {"yes" :1, "no" : 0}, "region": {"southeast": 0, "southwest": 1, "northwest" : 2, "northeast" : 3}})
# southeast --> 0
# southwest --> 1
# northwest --> 2
# northeast --> 3
```

dataset

## Split the dataset into the dependent and the independent variable.

```python
x = dataset.drop("charges", axis=1)
y = dataset["charges"]
```

## Training and testing Split.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
print(x_train, x_test)
```

## Linear Regression Model Implementation.

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
#  Predicting the x_train Value.
training_data_predict = model.predict(x_train)
```

```python
# R Squared Value  
error_rate = metrics.r2_score(training_data_predict , y_train)
print("R Squared Value.", error_rate * 100)
```

```python
# Visualize the Graph between the Actual and the predicted Value.
plt.figure(figsize=(7,7))
plt.scatter(y_train , training_data_predict)
plt.xlabel("Actual Data")
plt.ylabel("Prediction Data")
plt.title("Actual vs Predicted Value")
```

```python
#  Predicting the x_test Value.
testing_data_predict = model.predict(x_test)
```

```python
error_rate2 = metrics.r2_score(testing_data_predict , y_test)
print("The Error Rate of the Model Perf.", error_rate * 100)
```

```python
# Visualize the Graph between the Actual and the predicted Value.
plt.figure(figsize=(7,7))
plt.scatter(y_test , testing_data_predict)
plt.xlabel("Actual Data")
plt.ylabel("Prediction Data")
plt.title("Actual vs Predicted Value")
```
