# Credit Card Fraud Detection

```python
# import the required Library for building this project.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

### Data Collecting and Processing.

```python
dataset = pd.read_csv("creditcard.csv")
dataset
```

```python
#  Show the last 5 row of the dataset.
dataset.tail(5)
```

```python
#  Check the features information for the dataset.
dataset.info()
```

```python
#  How many missing values in each column.
dataset.isnull().sum()
```

```python
# Check the distinct value of the legit and the fraudelant
dataset["Class"].value_counts()

# There are the 492 Fraudent Columns.
```

This Data is Highly Unbalanced

0 -> Normal Transaction                
1 -> Fradulent Transaction

```python
# Separate the data for the analysis.
legit = dataset[dataset["Class"] == 0]
fraud = dataset[dataset["Class"] == 1]
```

```python
print(legit.shape, fraud.shape)
```

```python
# Statistical Measures of the Data.
legit["Amount"].describe()
```

```python
fraud["Amount"].describe()
```

```python
# Compare the values for both Transactions.
dataset.groupby("Class").mean()
```

## Under-Sampling Process
Build a Sample Dataset containing similar distribution of normal transactions and fradulent Transactions

492 -> Number of Fradulent Transactions                 
2L+ -> Number of Legit Transactions

```python
legit_sample = legit.sample(n = 492)
legit_sample
```

## Concatenating two DataFrames.

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
new_dataset
```

```python
new_dataset["Class"].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

## Splitting the data into Features and Targets.

```python
x = new_dataset.drop("Class", axis = 1)
y = new_dataset["Class"]
```

```python
print(x)
print(y)
```

## Splitting the data into the training and the testing Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

## Logistic Regression Model Training

```python
model = LogisticRegression(max_iter=500)
```

```python
model.fit(x_train, y_train)
```

## Logistic Regression Model Evaluation

```python
#  Find the accuracy Score for the training and the testing dataset.
training_data_pred = model.predict(x_train)
training_data_accu = accuracy_score(training_data_pred, y_train)
print("The Accuracy score for the training are:-", training_data_accu * 100)

testing_data_pred = model.predict(x_test)
testing_data_accu = accuracy_score(testing_data_pred, y_test)
print("The Accuracy score for the testing are:-", testing_data_accu * 100)
```
