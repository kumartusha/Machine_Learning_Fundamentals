# Regularization and its all types in Machine Learning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv(r"C:\Users\USER\OneDrive\Desktop\Housing.csv")
dataset.head(10)
```

```python
plt.figure(figsize= (10,10))
sns.heatmap(data=dataset.corr(), annot=True)
plt.show()
```

```python
# Input and the Output Data Variable (Independent or Dependent Variable )
x = dataset.iloc[:, :-1]
y = dataset["price"]
```

```python
sc = StandardScaler()
#  Using for fit the model..
sc.fit(x)
# Using for transform the model
x = pd.DataFrame(sc.transform(x), columns=x.columns)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Linear Regreesion, Lasso and Ridge

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd
```

# Linear Regression

```python
 lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test) * 100
```

```python
#  Check the Error like MSE, RMSE and all for the Linear Regression
 #  Mean Square Error..
print(mean_squared_error(y_test, lr.predict(x_test)))
 #  Mean Absolute Error..
print(mean_absolute_error(y_test, lr.predict(x_test)))
 #  Root Mean Absolute Error..
print(np.sqrt(mean_squared_error(y_test, lr.predict(x_test))))
```

```python
plt.figure(figsize=(15,5))
plt.bar(x.columns, lr.coef_)
plt.title("Linear Regression")
plt.xlabel("Columns")
plt.ylabel("Coef")
plt.show()
```

# Lasso Technique in Machine Learning

```python
la = Lasso(alpha = 0.5)
la.fit(x_train, y_train)
la.score(x_test, y_test)*100
```

```python
#  Check the Error like MSE, RMSE and all for the Lasso
 #  Mean Square Error..
print(mean_squared_error(y_test, la.predict(x_test)))
 #  Mean Absolute Error..
print(mean_absolute_error(y_test, la.predict(x_test)))
 #  Root Mean Absolute Error..
print(np.sqrt(mean_squared_error(y_test, la.predict(x_test))))
```

```python
plt.figure(figsize=(15,5))
plt.bar(x.columns, la.coef_)
plt.title("Lasso")
plt.xlabel("Columns")
plt.ylabel("Coef")
plt.show()
```

# Ridge Regularization in Machine Learning

```python
ri = Ridge(alpha = 10)
ri.fit(x_train, y_train)
ri.score(x_test, y_test) * 100
```

```python
#  Check the Error like MSE, RMSE and all for the Ridge
 #  Mean Square Error..
print(mean_squared_error(y_test, ri.predict(x_test)))
 #  Mean Absolute Error..
print(mean_absolute_error(y_test, ri.predict(x_test)))
 #  Root Mean Absolute Error..
print(np.sqrt(mean_squared_error(y_test, ri.predict(x_test))))
```

```python
plt.figure(figsize=(15,5))
plt.bar(x.columns, ri.coef_)
plt.title("Ridge")
plt.xlabel("Columns")
plt.ylabel("Coef")
plt.show()
```

```python
df = pd.DataFrame({"col_name": x.columns, "Linear Regression": lr.coef_, "Lasso": la.coef_, "Ridge" : ri.coef_})
df
```
