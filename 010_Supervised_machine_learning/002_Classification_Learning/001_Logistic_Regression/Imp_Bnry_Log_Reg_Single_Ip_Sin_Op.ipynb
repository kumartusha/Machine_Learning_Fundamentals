# Implementation of the Binary Logistic Regression ( Single Input and single Output)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"C:\Users\USER\OneDrive\Desktop\Github\Machine_Learning_Github\Machine_Learning_Fundamentals\010_Supervised_machine_learning\002_Classification_Learning\Social_Network_Ads.csv")
dataset.head(3)
```

```python
#  Predict the purchased based on the user Age
# Remove all the unnecessary columns from the dataset.
dataset.drop(columns=["User ID", "Gender", "EstimatedSalary"],inplace=True)
dataset
```

```python
#  First check the given data is applicable for the logistics regression or not..
plt.figure(figsize=(4,4))
sns.scatterplot(x = "Age", y = "Purchased", data = dataset)
plt.show()
```

```python
x = dataset[["Age"]]
y = dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
# Calculate the accuracy of the model.
lr.score(x_test, y_test) * 100
```

```python
#  Predict the value based on the input.
lr.predict([[32]])
```

```python
#  First check the given data is applicable for the logistics regression or not..
plt.figure(figsize=(15,5))
sns.scatterplot(x = "Age", y = "Purchased", data = dataset)
sns.lineplot(x = "Age", y = lr.predict(x), data = dataset, color = "red")
plt.show()
```
