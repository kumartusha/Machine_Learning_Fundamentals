```python
import pandas as pd
```

```python
dataset = pd.read_excel(r"C:\Users\USER\OneDrive\Desktop\Github\005_Machine_Learning_Github\Machine_Learning_Fundamentals\010_Supervised_machine_learning\002_Classification_Learning\008_K_Nearest_Neighbours/multiple_linear_regression_dataset.xlsx")
dataset.head(3)
```

```python
#  Check the null Value..
dataset.isnull().sum()
```

```python
x = dataset.drop(columns = "income")
y = dataset["income"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
knn.score(x_test, y_test) * 100
```

```python

#  find the train and test accuracy till 30

for i in range(1,30):
    knn1 = KNeighborsClassifier(n_neighbors=i)
    knn1.fit(x_train, y_train)
    print(i, knn1.score(x_train, y_train)*100, knn1.score(x_test, y_test)*100)
```
