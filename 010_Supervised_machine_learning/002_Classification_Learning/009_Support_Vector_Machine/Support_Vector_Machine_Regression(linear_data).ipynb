```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("placement_package.csv")
dataset
```

```python
# dataset.isnull().sum()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x = "cgpa", y = "package" , data = dataset)
plt.show()
```

```python
#  we need to apply the support Vector Machine Using the regression..
```

```python
x = dataset[["cgpa"]]
y = dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.2, random_state=42)
```

```python
#  We need to create the reference of the SVM.
from sklearn.svm import SVR
```

```python
sv = SVR(kernel='linear', degree=7)
sv.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test) * 100
```

```python
sv.score(x_train, y_train) * 100
```

```python
#  Make the Predicting Line..
sns.scatterplot(x = "cgpa", y = "package" , data = dataset)
plt.scatter(dataset["cgpa"], sv.predict(x), color = "red")
plt.show()
```
