# Implementation of the Binary Logistic Regression ( Multiple Input and single Output)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\USER\OneDrive\Desktop\Github\Machine_Learning_Github\Machine_Learning_Fundamentals\010_Supervised_machine_learning\002_Classification_Learning\placement.csv")
dataset
```

```python
#  Prediction of the placement based on the CGPA and the IQ level.
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(x = "cgpa", y = "iq", data = dataset, hue = "placement")
# plt.legend(loc = 1)
plt.show()
```

```python
x = dataset.iloc[:,: -1]
y = dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test) * 100
```

```python
lr.predict([[5.9, 106]])
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# How to build the decision boundary..
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(),clf = lr)
```
